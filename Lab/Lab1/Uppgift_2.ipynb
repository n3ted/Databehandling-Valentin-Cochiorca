```python
import pandas as pd
import plotly_express as px
```

```python
# a) Läs in data Covid 19 Vaccine 
file_path_vaccinations = '../Lab1/Data/Folkhalsomyndigheten_Covid19_Vaccine.xlsx'

# a) Läs bladet Vaccinerade kommun och ålder 
df = pd.read_excel(file_path_vaccinations, sheet_name='Vaccinerade kommun och ålder')
```

```python
# Räkna antal unika län i dataframe 
antal_län = len(df['Län'].unique())

# Skriv ut antal län
print(f'Total antal Län i datasetet: {antal_län}')
```

```python
# Räkna antal unika kommun i dataframet 
antal_kommuner = len(df['Kommun'].unique())

# Skriv ut antal kommuner 
print(f'Total antal kommuner i datasetet: {antal_kommuner}')
```

```python
# c) Beräkna totala befolkningen från datasetet
totala_befolkning = df['Befolkning'].sum()

# Skriv ut totala befolningen 
print(f'Total befolkning:{totala_befolkning: ,}')
```

```python
# d) Beräkna antal barn i Sverige baserad på datasetet

# Antal barn i datasetet under 18 år 
barn_data = df[df['Ålder'].isin(['12-15', '16-17'])]
antal_barn = barn_data['Befolkning'].sum()

# Beräkna antal barn i Sverige med dagens statistik 
procent_barn = antal_barn / totala_befolkning
statisktik_population_sverige = 10634798
estimerad_barn_sverige = statisktik_population_sverige * procent_barn

# Skriv ut antal barn under 18 år 
print(f'Estimerad barn under 18 år i Sverige: ca = {int(estimerad_barn_sverige): ,}')
```

```python
# e) Diagram med åldersfördelning i Sverige
fig = px.histogram(df, x='Ålder', y='Befolkning', color='Ålder')

# Vissa diagrammet 
fig
```

```python
# Gruppera data per län och beräkna medelvärde för doskategori
vaccin_data_per_län = df.groupby('Län_namn').agg(
    total_1_dos = pd.NamedAgg(column='Andel minst 1 dos', aggfunc='mean'),
    total_2_doser = pd.NamedAgg(column='Andel minst 2 doser', aggfunc='mean'),
    total_3_doser = pd.NamedAgg(column= 'Andel 3 doser', aggfunc='mean')
).reset_index()

# Skapa stapeldiagram med doskategori och län
fig = px.bar(vaccin_data_per_län, x='Län_namn',
    y=['total_1_dos', 'total_2_doser', 'total_3_doser'],
    title='Andel doser per län',
    labels={'value': 'Andel doser', 'variable': 'Dostyp'},
    barmode='stack')

# Visa diagrammet 
fig
```

```python
# f) Ta Västra Götalands län och Stockholms län och rita diagram över andelen

# Sortera data för Västara Götalands län och Stockholm län
valda_län = df[df['Län_namn'].isin(['Västra Götalands län', 'Stockholms län'])]

# Gruppera data per län och summera antal personner för varje doskategori
vaccine_data_valda_län = valda_län.groupby('Län_namn').agg(
    hela_befolkning = pd.NamedAgg(column='Befolkning', aggfunc='sum'), 
    total_1_Dos=pd.NamedAgg(column='Antal minst 1 dos', aggfunc='sum'),
    total_2_Doser=pd.NamedAgg(column='Antal minst 2 doser', aggfunc='sum'),
    total_3_Doser=pd.NamedAgg(column='Antal 3 doser', aggfunc='sum'),
    total_4_Doser=pd.NamedAgg(column='Antal 4 doser', aggfunc='sum')   
).reset_index()

# Beräkna andelen befolkning per län för varje doskategori
vaccine_data_valda_län['Andel_1_Dos'] =vaccine_data_valda_län['total_1_Dos'] / vaccine_data_valda_län['hela_befolkning']
vaccine_data_valda_län['Andel_2_Doser'] =vaccine_data_valda_län['total_2_Doser'] / vaccine_data_valda_län['hela_befolkning']
vaccine_data_valda_län['Andel_3_Doser'] =vaccine_data_valda_län['total_3_Doser'] / vaccine_data_valda_län['hela_befolkning']
vaccine_data_valda_län['Andel_4_Doser'] =vaccine_data_valda_län['total_4_Doser'] / vaccine_data_valda_län['hela_befolkning']

# Skapa stämpeldiagram för varje för varje doskategori och valda länen 
fig = px.bar(vaccine_data_valda_län, x='Län_namn', y=['Andel_1_Dos', 'Andel_2_Doser', 'Andel_3_Doser', 'Andel_4_Doser'],
title='Andel vaccinerade med minst 1, 2, 3, 4 doser i valda län', barmode='group')
# Byt namn på Y-axeln
fig.update_yaxes(title='Andel av befolkningen (%)')

# Vissa diagrammet 
fig
```

# Encoding categorical features

Categorical features are non-numerical features with a limited amount of posible options.

Before feeding them into most machine lerning algorithms we must convert them into numerical features using either:

- **Label encoding:** Each unique category is assigned to a numerical value: 0, 1, 2, 3..
- **One-hot encoding;** A new binary feature is created for each category.

```python
import pandas as pd 
import seaborn as sns 
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
cat_column = [col for col in titanic.columns if titanic[col].dtype in ['object', 'category'] ]
cat_column
```

```python
df_categories = titanic[cat_column]
df_categories
```

```python
for cat in cat_column:
    print(f'{cat} : {df_categories[cat].unique()}')
```

# Label Encoding 

Label encoding maps each category into a numerical value.

Use label encoding if:

- Catefories have a natural order 
- There are only 2 categories 
- Uding One-hot encoding leads to a large number of features 

```python
# Manual encoding 
titanic['embarked'] = titanic['embarked'].map({'S' : 0, 'C' : 1, 'Q' : 2})
titanic
```

# Auto-maping using Label Encoder from Scikit-learn

```python
# Use: pipenv install scikit-learn, to install package.
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
titanic['deck'] = le.fit_transform(titanic['deck'])
titanic
```

# One-hot encoding 

In one-hot encoding, a new binary feature is created for each category, and the values of that future is set to 1 if the observation belongs to that category, and 0 otherwise.

**Use one-hot** encoding if: 

- Categories have no natural order.
- Numbers of categories are small (but not 2)

```python
titanic.head()
# titanic.to_json("../Data/titanic_encoded.json", orient="records")
```

```python
titanic = pd.get_dummies(data=titanic, columns=['class'])
```

```python
titanic
```

# Auto-maping remaining non-numerical features 



```python
cat_columns = [col for col in titanic.columns if titanic[col].dtype in ['object', 'category', 'bool']]
cat_column
```

```python
for column in cat_columns:
    titanic[column] = le.fit_transform(titanic[column])

titanic
```

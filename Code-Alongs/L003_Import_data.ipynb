# Dataframe from file 

Pandas supports loading data from multiple file formats, including: excel, csv, json, xml

```python
import pandas as pd
```

```python
# Read csv file with pandas
students = pd.read_csv('../Data/students.csv', header=0)
students.head(1)
```

```python
# Read json file with pandas 
autos = pd.read_json('../Data/autos.json')

#Show part och data with head functon -> head() is set default to 5
autos.head()
```

```python
#Read excel data whit pandas 
scb = pd.read_excel('../Data/scb_pop.xlsx', skiprows=3)
scb.head()
```

```python
# Ways to show the data 
autos.head(10) # Nr to show from start 
autos.tail()  # Nr to show from end 
autos.sample(5) # Nr of samples from all data
```

```python
#Set options how the dataframe to be shown
pd.set_option('display.max_columns', None)
pd.set_option("display.max_rows", 6)
```

```python
# shape reference function returns a tuple with rows and col

# Save rows and col each in a variable 
rows, columns = autos.shape
print(f'Dataframe has {rows} rows and {columns} colums!')
```

```python
#Save nr of rows from dataframe using len function
rows = len(autos)

#Save nr of columns from datafram using len function 
columns = len(autos.columns)

print(f'Dataframe has {rows} rows and {columns} colums!')
```

# Convert to common formats

```python
# Copy dataframe to clipbord (then use ctrl+V)
autos.to_clipboard()
```

```python
autos.to_csv('../Data/autos.csv', index=None)
csv_string = autos.to_csv()
print(csv_string[:1000])
```

```python
autos.to_json('../Data/autos.json', orient='records')
json_string = autos.to_json(indent=4, orient='records')
print(json_string[:200])
```
